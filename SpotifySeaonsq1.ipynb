**Question 1**

    Does seasonality impact what genre of music people listen to throughout the year?


**Caveats**

    Spotify is a popular streaming company that currently holds 433 million active monthly users in 2022. Spotify owns 36% of the worldwide music streaming market, so it is the number one music streaming platform. North America comes in at the number 2 spot holding 22% of the share. Naturally as humans we like to play music in all type of settings. We love to curate playlist geared towards our moods and specific occasions. So we decided to dive into whether specific genres grow popularity in different seasons of the year. To keep the data cohesive with our Grammy statistics,  we decided to dive into Spotify's 2019 Top Weekly Songs in the US that were eligible for the 62nd Grammy Awards (2020). 


**Definitions**

    To create a sample size we choose the first week of top songs of a chosen month to represent a specific season.

    - January 2019 represents our Winter season
    - April 201 9represents our Spring season
    - July 2019 represents our Summer season
    -October 2018 represents our  Autumn season 

    Each Top Weekly Chart holds 200 songs, which brings our sample size to 800 songs and a grand total of 2,350,492,289 streams. 

**Methodology**

    1. We combined the seasonal dataset and were able to calculate the total numbers of streams using the "Streams" column for each genre. We turned this into a bar chart to show what genre's are the most popular across the data. 
 
    2. After looking at the data we decided to focus on the Top 5 genre's out of the grand total and break those "Streams" down to each season. 

    3. Lastly, we created a bar chart to how listening preferences are effected by seasonality. 



**Statisical Analysis**

    When we created the "Top 5 Genre's Per Season in 2019" chart, we automatically saw a drastic chnage with the "Pop" category. In Autumn the "Pop" category started at 7.68%,  reached its peak in the Summer with a 11.8% increase then made a total decline of 7% by Winter. 

    Across all seasons the" Rap/HipHop" genre reigns over the other genres but makes a huge decline from autumn to spring by 25%. 


**Conclusion**

    So coming back to our original question...Does seasonality impact what genre of music people listen to throughout the year? 

    Absolutely. 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
combineddata = "resources/combinedseasonsgenre.csv"


combinedseasondata = pd.read_csv(combineddata)

combinedseasondata
```

```python
combinedseasondata = combinedseasondata.rename(columns={"artist_names": "Artist", "track_name": "Track Name", 
                                        "rank": "Rank", "peak_rank": "Peak Rank", "previous_rank": "Previous Rank", 
                                       "weeks_on_chart": "Weeks on Chart", "streams": "Streams", "source": "Label", 
                                                       "streams": "Streams", "season": "Season", "month": "Month", "week": "Week",
                                                       "genre":"Genre"})

combinedseasondata
```

```python
seasonindex = combinedseasondata.set_index("Season")

seasonindex
```

```python
seasongroup = seasonindex.groupby(["Season", "Genre"])

seasongenre = seasongroup["Streams"].sum()

seasongenre

#This data set shows the total streams for each genre in each season 
```

```python
spotifygroup = seasonindex.groupby(["Genre"])
spotifystreams = spotifygroup["Streams"].sum()
spotifystreams

#Total number of streams for each Genre 

#here are the top 5 genre 
# Rap/HipHop           1329851564
# Pop                   314590239
# Alternative           295970285
# R&B/Soul              163200165
# Dance                 107331633
```

```python
total_streams = combinedseasondata["Streams"].sum()

total_streams

#Total number of streams for the entire sample data 
```

```python
total_percentage = spotifystreams/total_streams*100

total_percentage.round(2)

#Top 5 Genres 
#Rap_HipHop - 56.58% 
#Pop - 13.38%
#Alternative 12.59%
#R&B/Soul 6.64%
#Dance 4.57%
```

```python
figure1 = spotifystreams.plot(kind="bar", facecolor="blue", figsize=(10,6),
                                title="Total Streams For Each Genre in 2019",
                                xlabel="Genres",
                                ylabel="Streams in Billions")

plt.show
```

```python
spotifygroup = seasonindex.groupby(["Season"])
spotifystreams = spotifygroup["Streams"].sum()
spotifystreams
```

```python
pop_df = combinedseasondata.loc[combinedseasondata["Genre"] == "Pop", :]

pop_df
```

```python
popindex_df = pop_df.groupby(["Season"])

pop_streams = popindex_df["Streams"].sum()

pop_streams
```

```python
rap_df = combinedseasondata.loc[combinedseasondata["Genre"] == "Rap/HipHop", :]
rap_df
```

```python
rapindex_df = rap_df.groupby(["Season"])

rap_streams = rapindex_df["Streams"].sum()

rap_streams
```

```python
alternative_df = combinedseasondata.loc[combinedseasondata["Genre"] == "Alternative", :]

alternative_df 
```

```python
alternativeindex_df = alternative_df.groupby(["Season"])

alternative_streams = alternativeindex_df["Streams"].sum()

alternative_streams
```

```python
rsoul_df = combinedseasondata.loc[combinedseasondata["Genre"] == "R&B/Soul", :]
rsoul_df
```

```python
rsoulindex_df = rsoul_df.groupby(["Season"])

rsoul_streams = rsoulindex_df["Streams"].sum()

rsoul_streams
```

```python
dance_df = combinedseasondata.loc[combinedseasondata["Genre"] == "Dance", :]
dance_df
```

```python
danceindex_df = dance_df.groupby(["Season"])

dance_streams = danceindex_df["Streams"].sum()

dance_streams
```

```python
top5genre_season_df = pd.DataFrame({"Rap/HipHop": rap_streams, 
                                 "Alternative": alternative_streams,
                                 "Pop": pop_streams, 
                                 "R&B/Soul": rsoul_streams, 
                                "Dance": dance_streams})

top5genre_season_df
```

```python
top5genre_perseason_df = top5genre_season_df

stream_plot = top5genre_perseason_df.plot(kind="bar", figsize=(20,3.5)).legend(loc='best')
plt.ylabel("Streams in Millions")
plt.xlabel("Seasons")
plt.title("Top 5 Genre's Per Season in 2019")


# plt.savefig("resources/most_streamed_entire_period_q3.png", bbox_inches="tight")
plt.show()
```
